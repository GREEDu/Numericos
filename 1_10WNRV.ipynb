```python
import h5py
import matplotlib.pyplot as plt


def graficar_hd5(datoscolegahd5):

    with h5py.File('datoscolega.hd5', 'r') as file:
        x = file['x'][:]
        y = file['y'][:]
        e = file['e'][:]

    plt.errorbar(x, y, e, fmt='o', ecolor='black', capsize=3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Datos con barras de error')
    plt.grid(True)
    plt.show()


if __name__ == '__main__':
    graficar_hd5('datoscolegahd5.hd5')

#la funcion abre un archivo .h5 basandose en el repositorio github, asi lee los tres datasets que se solicitan
#el grafico contiene los ejes y a su vez usa e como barra de error, el codigo, no lograr√° correr porque falta el archivo que se supone debe leer
#pero el codigo seria funcional si tuvieramos realmente los datos del colega en un archivo HDF5
```

```python
# ============================================
# Tarea — valores minimos usando jax
# ============================================

import jax
import jax.numpy as jnp
import numpy as np
jax.config.update("jax_enable_x64", True)

# ============================================
# 1. Funcion
# ============================================
def volumen(theta):
    theta1, theta2 = theta
    return jnp.sinh(theta1**4 + theta2**2) - 1.0

# Gradiente con autodiferenciación
grad_volumen = jax.grad(volumen)

# ============================================
# 2. Descenso por gradiente
# ============================================
def descenso_gradiente(theta_init, lr=0.01, iteraciones=2000):
    theta = theta_init
    historial = []
    trayectoria = []

    for i in range(iteraciones):
        grad = grad_volumen(theta)
        theta = theta - lr * grad

        # Proyección al primer cuadrante
        theta = jnp.maximum(theta, 0.0)

        historial.append(volumen(theta))
        trayectoria.append(theta)

    return theta, jnp.array(historial), jnp.array(trayectoria)


# ============================================
# 3. Minimización
# ============================================
theta_inicial = jnp.array([1.0, 1.0])
theta_opt, hist, trayectoria = descenso_gradiente(theta_inicial, lr=0.01, iteraciones=2000)

print("Theta inicial:", theta_inicial)
print("Theta óptimo encontrado:", theta_opt)
print("Volumen mínimo aproximado:", volumen(theta_opt))
print("Mínimo teórico: (0,0) → V = -1")

#queda demostrado que el Volumen es igual a 'sinh(theta1**4 + theta2**2) - 1.0'
#alcanzando su minimo en theta 1 y 2 iguales a 0, con un minimo muy cercano a -1
#el descenso con jax va actualizando theta, y numericamente se obitne un punto optimo cercano a cero
#dando asi el valor aproximado cercano a -1 coincidiendo con el valor teorico
#esta diferencia se debe al metodo usado por pasos finitos, y un numero finito de iteraciones
#cito a un compañero de magister que me ayudó en un principio cuando no lograba instalar bien jax, teniendo que
# usar jupyter para reemplazar vsc,pero luego sin poder correrlo de buena manera
# y tambien cito al profesor que finalmente me sugirió correrlo en collab, facilitando el trabajo


#deje como comentarios las primeras celdas con los comandos de instalaciones que estuve usando
```
